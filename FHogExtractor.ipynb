```python
l1 = []
PATH = os.getcwd()
data_path = PATH + '/data'
data_dir_list = os.listdir(data_path)

img_data=[]
for dataset in data_dir_list:
    img_list=os.listdir(data_path+'/'+ dataset)
    print ('Loaded the images of dataset-'+'{}\n'.format(dataset))
    for img in img_list:
        #img1 = (imageio.imread(img).astype(np.float64)/255.0)
        input_img=cv2.imread(data_path + '/'+ dataset + '/'+ img )
        im1 = cv2.resize(input_img.astype(np.float64)/255.0, (150, 150))
        hog1 = pyhog.features_pedro(im1, 30)
        a = np.array(hog1)
        b = a.flatten()
        l = b.tolist()
        #print l
        l1.append(l)

image_features_arr = np.array(l1)
print (image_features_arr.shape)
np.savetxt('feature_vectors.txt',image_features_arr)
    
```
